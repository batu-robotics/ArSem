## 2. Dosyayı Çağırma

# 3. Eksik Veriyi Yönetme

```python
veri_4=veri.fillna(veri.mean())
veri_4.describe()
```

# 4. Price için aykırı değer analizi

```python
q_low=veri_4["price"].quantile(0.05)
q_high=veri_4['price'].quantile(0.95)
sns.boxplot(data=veri_4)

veri_yeni=veri_4[(veri_4['price']>q_low)&(veri_4['price']<q_high)]
veri_yeni.count()
```

# 5. Öznitelik Analizi

```python
bagimli=veri_yeni.iloc[:,-1]
bagimsiz=veri_yeni.iloc[:,:-1]
om=bagimsiz.corr()
sns.heatmap(om,annot=True)
```

# Kategorik Verinin Kodlanması

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

le.fit(veri_yeni['make'])
veri_yeni['make']=le.transform(veri_yeni['make'])
```

```python
le=LabelEncoder()

for i in range(veri_yeni.shape[1]):
  if veri_yeni.iloc[:,i].dtype=='object':
    le.fit(veri_yeni.iloc[:,i])
    veri_yeni.iloc[:,i]=le.transform(veri_yeni.iloc[:,i])

sns.heatmap(veri_yeni.corr(),annot=True)
```
